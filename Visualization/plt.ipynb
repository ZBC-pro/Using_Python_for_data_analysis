```python
from pathlib import Path
import csv
import pandas as pd

from matplotlib import pyplot as plt

df = pd.read_csv('week_plt.csv')

categories = df.iloc[:, 0]
data = df.iloc[:, 1:]

for i, category in enumerate(categories):
    plt.plot(data.columns, data.iloc[i], marker='o', label=category)

plt.title('Line chart')
plt.xlabel('type')
plt.ylabel('value')
plt.legend(title='categories')

plt.show()

    
        
```
