```python
import os
import requests
import pandas as pd
from urllib.parse import urlparse

# 定义路径变量
excel_file_path = 'test_image.xlsx'  # 替换为你的 Excel 文件路径
output_folder = './output'  # 输出文件夹路径
sheet_names = ['纸币1', '纸币2']

# 确保输出文件夹存在
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍历每一个工作表
for sheet_name in sheet_names:
    print(f"处理工作表: {sheet_name}")
    
    # 读取指定的工作表
    df = pd.read_excel(excel_file_path, sheet_name=sheet_name)
    
    # 遍历每一行，按第一列创建文件夹
    for index, row in df.iterrows():
        folder_name = str(row[0])
        folder_path = os.path.join(output_folder, folder_name)
        positive_folder = os.path.join(folder_path, '正面')
        negative_folder = os.path.join(folder_path, '反面')
        
        # 确保文件夹和子文件夹存在
        if not os.path.exists(positive_folder):
            os.makedirs(positive_folder)
        if not os.path.exists(negative_folder):
            os.makedirs(negative_folder)
        
        # 处理正面图片链接（第九列）
        if len(row) >= 9 and pd.notna(row[8]):
            url = str(row[8])
            if url.startswith(('http://', 'https://')):
                try:
                    parsed_url = urlparse(url)
                    file_name = os.path.basename(parsed_url.path)
                    file_path = os.path.join(positive_folder, file_name)
                    
                    # 下载图片
                    response = requests.get(url)
                    #print(f"下载正面图片 URL: {url}, 状态码: {response.status_code}")
                    response.raise_for_status()  # 确保请求成功
                    
                    # 保存图片
                    with open(file_path, 'wb') as file:
                        file.write(response.content)
                    
                    print(f"已下载正面图片: {file_path}")

                except requests.RequestException as e:
                    print(f"下载正面图片失败: {url}，原因: {e}")
        
        # 处理反面图片链接（第十列）
        if len(row) >= 10 and pd.notna(row[9]):
            url = str(row[9])
            if url.startswith(('http://', 'https://')):
                try:
                    parsed_url = urlparse(url)
                    file_name = os.path.basename(parsed_url.path)
                    file_path = os.path.join(negative_folder, file_name)
                    
                    # 下载图片
                    response = requests.get(url)
                    #print(f"下载反面图片 URL: {url}, 状态码: {response.status_code}")
                    response.raise_for_status()  # 确保请求成功
                    
                    # 保存图片
                    with open(file_path, 'wb') as file:
                        file.write(response.content)
                    
                    print(f"已下载反面图片: {file_path}")

                except requests.RequestException as e:
                    print(f"下载反面图片失败: {url}，原因: {e}")

print("所有图片下载完成！")
```
